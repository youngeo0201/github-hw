```python
#39
def print_odd_numbers_39(max_num):
    if max_num < 0:
        print("양수를 입력해야 합니다.")
        return

    print(f"결과: 0부터 {max_num}까지의 홀수:")
    for i in range(max_num + 1):
        if i % 2 != 0:
            print(i, end=' ')
    print()
try:
    user_input_39 = int(input("양수를 입력하세요: "))
    print_odd_numbers_39(user_input_39)
except ValueError:
    print(" 숫자를 입력해주세요.")
```

```python
#41
def calculate_min_max_41(n1, n2, n3, n4):
    numbers = [n1, n2, n3, n4]
    return max(numbers), min(numbers)
try:
    print("숫자 4개를 띄어쓰기로 구분하여 입력하세요: ")
    scores_41 = list(map(int, input().split()))

    print("결과:")
    if len(scores_41) == 4:
        maximum, minimum = calculate_min_max_41(*scores_41)
        print(f"입력 숫자: {scores_41}")
        print(f"최댓값: {maximum}")
        print(f"최솟값: {minimum}")
    else:
        print("숫자 4개를 정확히 입력해야 합니다.")

except ValueError:
    print("유효한 숫자만 입력해야 합니다.")
except Exception as e:
    print(f" 오류 발생: {e}")
```

```python
#43
def calculate_factorial_43(n):
    if n == 0:
        return 1
    elif n < 0:
        return "0보다 크거나 같은 정수를 입력해야 합니다."
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result
try:
    user_input_43 = int(input("0보다 크거나 같은 정수 n을 입력하세요: "))
    factorial_result = calculate_factorial_43(user_input_43)
    print(f"결과: {user_input_43}! = {factorial_result}")
except ValueError:
    print("유효한 정수를 입력해주세요.")
```

```python
#44
def calculate_sum_over_30_44(start_i, start_j):

    total_sum = 0
    print(f"결과: i={start_i}부터, j={start_j}부터 시작")

    for i in range(2, 10):
        for j in range(2, 10):
            if i < start_i or j < start_j:
                continue

            product = i * j
            if product >= 30:
                print(f"  {i} * {j} = {product} ", end='')
                total_sum += product

    print(f" 합: {total_sum}")
    return total_sum

try:
    user_input_44 = input("2이상 9이하의 양수 2개를 띄어쓰기로 구분하여 입력하세요: ").split()
    if len(user_input_44) == 2:
        i_start = int(user_input_44[0])
        j_start = int(user_input_44[1])

        if 2 <= i_start <= 9 and 2 <= j_start <= 9:
            calculate_sum_over_30_44(i_start, j_start)
        else:
            print(" i와 j는 반드시 2이상 9이하의 양수여야 합니다.")
    else:
        print("숫자를 2개 입력해야 합니다.")
except ValueError:
    print("유효한 정수만 입력해야 합니다.")
```

```python
#45
def calculate_cumulative_sum_45(input_list):
    cumulative_sum = 0
    cumulative_sums_list = []

    for element in input_list:
        cumulative_sum += element
        cumulative_sums_list.append(cumulative_sum)

    print("결과:")
    print(f"입력 리스트: {input_list}")
    print(f"누적 리스트: {cumulative_sums_list}")
    print(f"최종 합: {cumulative_sum}")
    return cumulative_sums_list

list_a = [1, 2, 3, 4, 5]
calculate_cumulative_sum_45(list_a)
```
