```python
#과제20
input_data = input("두 정수(시작 N, 끝 M)를 공백으로 구분하여 입력하세요 (N < M): ")
start_n, end_m = map(int, input_data.split())

power_of_two_list = [2**i for i in range(start_n, end_m + 1)]

list_len = len(power_of_two_list)

if list_len >= 4:
    result_list = power_of_two_list[:1] + power_of_two_list[2:-2] + power_of_two_list[-1:]

elif list_len == 3:
    result_list = [power_of_two_list[0], power_of_two_list[-1]]

else:
    result_list = []

print(f"실행결과: {result_list}")
```
